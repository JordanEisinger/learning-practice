```mojo
# Python

import math

inside = 0
resolution = 100000

for x in range(resolution):
    x = x/resolution
    for y in range(resolution):
        y = y/resolution
        if math.sqrt((x*x)+(y*y)) <= 1:
            inside += 1

print("pi equals: ", 4*(inside / (resolution ** 2)))
```
