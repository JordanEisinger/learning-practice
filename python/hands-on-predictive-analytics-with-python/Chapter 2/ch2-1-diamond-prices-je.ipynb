```python
import numpy as np
import pandas as pd
import os
```

# Data Import

```python
DATA_DIR = '../test-data'
FILE_NAME = 'diamonds.csv'
data_path = os.path.join(DATA_DIR, FILE_NAME)
print(data_path)
```

```python
# Define DataFrame
diamonds = pd.read_csv(data_path)
print(diamonds.shape)
diamonds.head()
```

```python
# Get summary statistics
diamonds.describe()
```

```python
# Check for zeros
diamonds.loc[diamonds['x']==0]
```

```python
# x=0
print(diamonds.loc[11182])
print()

# Replace x=0 with the median value
diamonds.loc[11182, 'x'] = diamonds['x'].median()
print(diamonds.loc[11182])
```

```python
# Only show rows where x and y are greater than 0
diamonds = diamonds.loc[(diamonds['x']>0) | (diamonds['y']>0)]
display(diamonds)
```
