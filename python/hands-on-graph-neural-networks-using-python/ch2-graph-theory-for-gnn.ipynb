Types of Graphs
* **Tree**: a connected, undirected graph with no cycles
* **Rooted Tree**: tree in which one node (or vertice) is designated as the root, and all other vertices are connected to it by a unique path
* **Directed Acyclic Graph (DAG)**: a directed graph that has no cycles
* **Biparite Graph**: a graph in which the vertices can be divided into two disjoint setsm such that all edges connect vertices in different sets
* **Complete Graph**: a graph in which every pair of vertices is connected by an edge

### Graph Algorithms

Breadth-First Searc (BFS)
* Starts at the root node and explores all the neighboring nodes at a particular level before moving to the next level
* Variations:
    * bidirectional BFS
    * A* Search

```python
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])

def bfs(graph, node):
    visited, queue = [node], [node]

    while queue:
        node = queue.pop(0)
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)

    return visited

bfs(G, 'A')
```

Depth-First Search (DFS)
* A recursive algorithm that starts at the root node and explores as far as possible along each branch before backtracking
* Notes:
    * Uses less memory than BFS, but doesn't guarantee the shallowest path
    * Can be trapped in infinite loops using DFS

```python
visited = []

def dfs(visited, graph, node):
    if node not in visited:
        visited.append(node)
        for neighbor in graph[node]:
            visited = dfs(visited, graph, neighbor)

    return visited

dfs(visited, G, 'A')
```
